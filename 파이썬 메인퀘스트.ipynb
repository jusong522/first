```python
import random


def make_account_num():
    num1 = str(random.randint(0, 999)).zfill(3)
    num2 = str(random.randint(0, 99)).zfill(2)
    num3 = str(random.randint(0, 999999)).zfill(6)
    return num1 + "-" + num2 + "-" + num3


class Account:
    account_cnt = 0

    def __init__(self, name, money):
        self.bank = "SC은행"
        self.name = name
        self.balance = money
        self.account = make_account_num()

        self.deposit_cnt = 0
        self.deposit_log = []
        self.withdraw_log = []

        Account.account_cnt += 1

    @classmethod
    def get_account_num(cls):
        print("총 계좌 개수:", cls.account_cnt)

    def deposit(self, money):
        if money <= 0:
            print("1원 이상 입금하세요.")
            return

        self.balance += money
        self.deposit_cnt += 1
        self.deposit_log.append(money)

        if self.deposit_cnt % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            print("이자 지급:", interest, "원")

    def withdraw(self, money):
        if money > self.balance:
            print("잔액이 부족합니다.")
            return

        self.balance -= money
        self.withdraw_log.append(money)

    def display_info(self):
        print(
            "은행:", self.bank,
            "| 이름:", self.name,
            "| 계좌:", self.account,
            "| 잔고:", format(self.balance, ","), "원"
        )

    def deposit_history(self):
        print(self.name, "입금 내역")
        if len(self.deposit_log) == 0:
            print("내역 없음")
        else:
            for i in range(len(self.deposit_log)):
                print(i + 1, ":", format(self.deposit_log[i], ","), "원")

    def withdraw_history(self):
        print(self.name, "출금 내역")
        if len(self.withdraw_log) == 0:
            print("내역 없음")
        else:
            for i in range(len(self.withdraw_log)):
                print(i + 1, ":", format(self.withdraw_log[i], ","), "원")


acc_list = []
acc_list.append(Account("파이썬", 10000))
acc_list.append(Account("리눅스", 500000))
acc_list.append(Account("판다스", 1500000))

Account.get_account_num()

acc_list[0].deposit(1000)
acc_list[0].deposit(1000)
acc_list[0].deposit(1000)
acc_list[0].deposit(1000)
acc_list[0].deposit(1000)
acc_list[0].withdraw(3000)

print("\n잔고 100만원 이상 고객")
for acc in acc_list:
    if acc.balance >= 1000000:
        acc.display_info()

print()
acc_list[0].deposit_history()
acc_list[0].withdraw_history()
```
